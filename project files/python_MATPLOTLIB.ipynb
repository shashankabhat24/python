### Bar charts


```python
labels = ['A','B','C']

plt.figure(figsize=(6,4),dpi=100)

plt.xlabel("x-axis")
plt.ylabel("y-axis")

bars=plt.bar(labels,values)
patterns=['*','/','o']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[1].set_hatch('*')
#bars[0].set_hatch('/')
#bars[2].set_hatch('o')

plt.show()
```
